# Aula 1 - Biblioteca Matplotlib

## Importações

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/imigrantes_canada.csv')
df.head()
```

```python
type(df)
```

```python
df.info()
```

```python
df
```

## Mudando o índice do DataFrame, alterando sua estrutura original (inplace=True)

```python
df.set_index('País',inplace=True)
```

```python
anos = list(map(str, range(1980,2014)))
```

```python
anos
```

```python
type(anos)
```

## Extrair a Série de dados do Brasil (loc())

- loc [] - pesquisa pelo label
- iloc [] - pesquisa pelo indice

```python
# Extrair a série de dados do Brasil
brasil = df.loc['Brasil',anos]
```

```python
brasil
```

```python
type(brasil)
```

```python
brasil_dict = {'ano':brasil.index.tolist(),
               'imigrantes':brasil.values.tolist()}
```

```python
df_brasil = pd.DataFrame(brasil_dict)
```

```python
df_brasil
```

## Criando o gráfico de linhas

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.grid()
#plt.scatter()
plt.show()
```

## Transformar o gráfico como figura

- Aumenta o potencial dos comandos

```python
# figsize=(largura(polegadas),altura(polegadas))
# 1 polegada == 2.54cm
# DPI - Dots per Inch, ou pontos por polegada
# Padrão Matplotlib -> DPI = 100
# Largura : 8 polegadas X 100 DPI = 800 pixels
# Altura : 4 polegadas X 100 DPI = 400 pixels

plt.figure(figsize=(8,4))
plt.plot(df_brasil['ano'], df_brasil['imigrantes'])
plt.title("Imigração de brasileiros para o Canadá")
plt.xlabel("Anos")
plt.ylabel("Número de Imigrantes")
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.show()
```

## Criando Subplots e adicionando rótulos

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(df_brasil['ano'],df_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')
plt.grid()
plt.show()
```

## Gráfico de linha e boxplot em subplots

```python
fig, axs = plt.subplots(1,2,figsize=(12,6))
# Subplot 1- Gráfico Linha
axs[0].plot(df_brasil['ano'],df_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].grid()
# Subplot 2 - Gráfico Boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('Box-plot da imigração do Brasil para o Canadá \n 1980 a 2013')
axs[1].set_xlabel('Ano')
axs[1].set_ylabel('Número de Imigrantes')
axs[1].grid()
```

## Comparativo dos maiores paíse da América Latina (Imigração Canadá)

```python
# Criar uma figura
fig,axs = plt.subplots(2,2,figsize=(10,6))

# Brasil
axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('Brasil')

# Colômbia
axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Colômbia')

# Argentina
axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

# Peru
axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')

fig.subplots_adjust(hspace=0.5, wspace=0.6)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de Imigrantes')
    ax.grid()
    
'''
    Usando ravel() - Achatar a matriz evitando o uso complexo do for
    Usando set_ylim() - Definir o limite dos valores
'''    
# Determinar o mínimo
ymin = 0

# Determinar o máximo
ymax = 7000

# Criar um for para usar ravel() - Achatar a matriz Bidimensional para um Unidimensional
for ax in axs.ravel():
    # usar set_ylim() - Determinar o mínimo eo máximo do eixo y
    # ax,set_ylim(ymin,ymax)
    ax.set_ylim(0,7000)
```
